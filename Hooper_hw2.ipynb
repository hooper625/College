```python
#CS620-Introduction to Data Science and Analytics
# Homework Exercise 2 (HW2): Fall 2024
# author: <Hooper, UIN>
import pandas as pd
import numpy as np

#Q1: Load the diamonds.csv file into a DataFrame
csv_file = 'https://www.cs.odu.edu/~sampath/courses/data/diamonds.csv'
df = pd.read_csv(csv_file)
df.head()

#Q2: List the data types of the DataFrame and the last 10 rows
print("Data types of the DataFrame:")
print(df.dtypes)
print("\nLast 10 rows of the DataFrame:")
print(df.tail(10))

#Q3: Convert the 'price' column to a NumPy array
diamond_price = df['price'].to_numpy()
print(f'\nMiddle diamond price/value:', {np.median(diamond_price)})
print(f'\nThe data in ascending order:\n{np.sort(diamond_price)}')
print(f'\nThe range of the data is: ',{(np.max(diamond_price) - np.min(diamond_price))})

#Q4: Show the “Premium” type diamonds’ depth, price, and clarity
premium_diamonds = df[df['cut'] == 'Premium']
print("\nThe Premium diamonds are:")
print(premium_diamonds[['depth', 'price', 'clarity']])

#Q5: Add a new column 'priceVolatility' as 12% of the 'price' column using a lambda function
print("\nThe new column 'priceVolatility' with color and carat are:")
df['priceVolatility'] = df['price'].apply(lambda x: x* 0.12)
print(df.loc[[4, 27, 1221], ['carat', 'color', 'priceVolatility']])

#Q6: Display the summary statistics for clarity “VS1”
print("\nSummary statistics for clarity 'VS1':")
clarity_vs1 = df[df['clarity'] == 'VS1']
print(clarity_vs1.describe())

#Q7: Add a new column called quality
#Q8: Fill the missing values of column “quality” by “NaN”
def classify_quality(depth):
    if 54 <= depth <= 57:
        return 'round diamond'
    elif 68 <= depth <= 74:
        return 'princecut diamond'
    else:
        return np.NaN
df['quality'] = df['depth'].apply(classify_quality)
print("\nThe new column 'quality' is:")
print(df[['depth', 'quality']])

#Q9: Make a copy of the data frame “diamond” called “diamondQ”...
diamondQ = df.copy()
diamondQ.dropna(inplace=True)
print("\nThe copy of the data frame 'diamond' is:")
print(diamondQ)
```
