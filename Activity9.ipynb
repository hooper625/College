```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from URL
df = pd.read_csv('https://www.cs.odu.edu/~sampath/courses/data/brfss.csv')

# Clean the dataset by dropping 'sex' column and NaN values
def cleanBRFSSFrame():
  df.drop('sex', axis=1, inplace=True)
  df.dropna(inplace=True)

cleanBRFSSFrame()

# Use Pandas' describe() method to display summary statistics
print("Summary Statistics for 'weight2':")
print(df['weight2'].describe())
# Display age quantiles (including the 70th percentile)
quantile_age = df['age'].quantile([0.25, 0.50, 0.70])
print("\nQuantiles for 'age' (25th, 50th, 70th):")
print(quantile_age)
# Create a violin plot for the 'htm3' (height in meters) column
sns.violinplot(y='htm3', data=df)
plt.title("Violin Plot of Height (htm3)")
plt.show()




```
